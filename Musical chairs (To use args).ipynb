# Musical chairs

How to play...
- Number of players has no limit.
- Each round start and end with a music
- When a round starts, chair will be placed in a circle with one less chair
- each participents need to occupy one chair when round ends
- repeat from point 3
- Who is remained will be crowned as winner
Game rules...
- no fighting

Instructions to use the python program
- enter the list of participants in as strings seperated by comma
- click ENTER to start the round
- click STOP button of Jupyter to end the round
Rest, is taken care

```python
# Takes string which will repeat for a given wait(in seconds) intervel
def timered(string, wait):
    import time
    while True:
        try:
            print (string)
            time.sleep(wait)
        except KeyboardInterrupt:
            break    

# Takes a string which will return in random colors(round robin fashion of RGB)
def colored(string):
    from colorama import Fore, Back              #To change the text color
    string_out = []
    in_red = lambda x : Fore.RED+x                      #Using lambda for red
    in_green = lambda x : Fore.GREEN+x                  #Using lambda for Green
    in_blue = lambda x : Fore.BLUE+x                     #Using lambda for Blue
    for index, letter in enumerate(string):
        if index%3 == 0:
            string_out.append(in_red(letter))
        elif index%2 == 0:
            string_out.append(in_green(letter))
        else:
            string_out.append(in_blue(letter))
    return ("".join(string_out))

def musical_chair(*members):        #using *args concept
    
    #library import
    import random                                       #To select player randomly
    import time                                         #To provide game time
    from colorama import Fore, Back, Style              #To change the text color
    
    #Declaratives
    members = list(members)                             #converting args-tuple to list
    startTime = time.time()                             #to store the start time of the game
    results = []                                        #Later use, To print final left overs of game
    in_white_black = lambda x : Fore.WHITE+Back.BLACK+x #To print white in black background
    in_red = lambda x : Fore.RED+x                      #Using lambda for red
    in_green = lambda x : Fore.GREEN+x                  #Using lambda for Green
    in_blue = lambda x : Fore.BLUE+x                    #Using lambda for Blue
    in_yellow_red = lambda x : Fore.YELLOW+Back.RED     #To print yellow in red background
    reset_color = lambda x : Style.RESET_ALL+x          #To reset colors
    
    #Actual game prgram starts here
    print(str(u"\U0001FA91")*10+ "  "+colored("MUSICAL CHAIRS")+ "  "+str(u"\U0001FA91")*10)         #To print game name in colors
    print("\nLets welcome our participants...."+"\n\n"+u'\U0001f60e'*22+"\n")
    for member in members:
        print(""+u"\U0001FA91"+member)                  #Unicodes available online, starts with u''
    print("\n\nRemember the controls\n\033[1mpress "+in_white_black("ENTER")+"\033[0;0m to start when asked to...\n\033[1mSTOP "+in_white_black("JUPYTER")+"\033[0;0m(stop kernal) to stop the round...")
    while True:
        start = input("""Lets remove one chair...
        and start the round by pressing \033[1m"""+in_white_black("ENTER")+"""\033[0;0m  \n""") #Multiline input
        if start == "":
            timered("Go "+in_blue("round")+reset_color(" and ")+in_red("round")+reset_color(" and ")+in_green("round")+"....\n", 2)
            # here - use multithreding
            
            print("\n...............STOP............")
            out = members.pop(members.index(random.choice(members)))     #logs the loosed player
            results.append(out)
            if len(members) == 1:
                print("\nThats a fighting and....")
                time.sleep(2)
                print(members[0]+" you are almost there.. push.. push.. and")
                time.sleep(5)
                print(out+"  "+u"\U0001F625")
                print("\nYaaa...... "+u"\U0001F44F"*5)
                print("    "+u"\U0001F451")
                print(members[0])
                print(u"\U0001F44F"*10)
                endTime = time.time() #to store the start time of the game
                print("\nTotal game time is "+str(int(endTime-startTime))+" seconds")
                print("\nHere are the order of left-overs\n")
                for i,m in enumerate(results):                      ##### Enumerator for index, value of a list
                    print (m+" left in round "+str(i+1))
                break
            else:
                print(colored("\nOOOOOOOHHHHHHHHHHH>>>>>>>>>>>>>\n"))
                for member in members:
                    print(u"\U0001FA91"+member+	u'\U0001f61b')
                print(out+"  "+u'\U0001f615')
                out = ""
    print("\n\n\n"+u"\U0001F64F"*30)
       
```
